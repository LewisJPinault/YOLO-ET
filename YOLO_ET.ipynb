```python
import turicreate
import matplotlib.pyplot as plt
from matplotlib import image
import matplotlib.patches as patches
import pandas as pd
import numpy as np

plt.rcParams.update({'font.size': 22}) #set font size
plt.rcParams['figure.figsize'] = [20, 10] #change these values to change image size
```

```python
# set up test data
test_images = tc.load_images('test_images')
test_annotations = tc.SFrame.read_json('labels_test.json')

test_images['annotations'] = test_annotations['annotations']
```

```python
##load model
model = tc.load_model("YOLO_ET")
```

```python
#apply model to test data
predictions = model.predict(test_images)
```

```python
# function to plot predicted boxes and/or ground truth boxes
def plot_image(impath, plot_box, annotations, plot_conf=False, gt=False, gtannotations=None):
    imdata = image.imread(impath)
    plt.imshow(imdata)

    if plot_box:
        if len(annotations) > 0:
            for i in range(len(annotations)):
                height = annotations[i]['coordinates']['height']
                width = annotations[i]['coordinates']['width']
                x = annotations[i]['coordinates']['x']
                y = annotations[i]['coordinates']['y']
                if plot_conf:
                    label = "%.1f" % (annotations[i]['confidence']*100) +'% '+annotations[i]['label']
                else:
                    label = annotations[i]['label']

                #create box
                rect = patches.Rectangle((x-width/2,y-height/2), width, height, linewidth=3, edgecolor='r', facecolor='none')

                plt.gca().add_patch(rect)

                plt.text(x - width/2, y + height/2+25,  # Adjust for label placement
                        label,  # Label text
                        color='r',  # Label color
                        ha='left',  # Alignment 'center', 'left', 'right'
                        fontsize=20  # Label font size
                        )
    if gt:
        #ground truth box
        if len(gtannotations) > 0:
            for i in range(len(gtannotations)):
                h2 = gtannotations[i]['coordinates']['height']
                w2 = gtannotations[i]['coordinates']['width']
                x2 = gtannotations[i]['coordinates']['x']
                y2 = gtannotations[i]['coordinates']['y']

                label2 =  gtannotations[i]['label']

                #create box
                rect2 = patches.Rectangle((x2-w2/2,y2-h2/2), w2, h2, linewidth=2, linestyle='--',edgecolor='b', facecolor='none')


                plt.gca().add_patch(rect2)
                plt.text(x2 - w2/2, y2 - h2/2-10,  # Adjust for label placement
                        label2,  # Label text
                        color='b',  # Label color
                        ha='left',  # Alignment 'center', 'left', 'right'
                        fontsize=20  # Label font size
                        )
```

```python
i=0 #plot i-th test_image
plot_image(test_images['path'][i],
           plot_box=True,
           annotations=predictions[i],
           plot_conf=True,
           gt=False)
```
